```python
# Import necessary libraries
from bs4 import BeautifulSoup  # BeautifulSoup is a library for parsing HTML and XML documents
import re  # The re module provides support for regular expressions

# Read the HOCR file
with open("/content/combined_hocr_141813.hocr", "r", encoding="utf-8") as file:
    hocr_content = file.read()  # The re module provides support for regular expressions

# Parse the HOCR content using BeautifulSoup
soup = BeautifulSoup(hocr_content, "html.parser")  # Create a BeautifulSoup object from the HOCR content

# Extract key information from the HOCR content

# Extract the title text from bounding box "bbox 270 119 854 259"
div_title = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 270 119 854 259" in value)

# Extract text from the first div
if div_title:
    title = div_title.get_text(strip=True)   # Extract text from the found div
else:
    title = ""  # Set title_text to empty string if the div is not found

# Extract the place names from bounding box "bbox 111 1497 978 1538"
div_place_tag = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 111 1497 978 1538" in value)

# Extract text from the third div with the specified format (\b[A-Z][a-z]+\b)
if div_place_tag:
    place_text = ""
    text = div_place_tag.get_text(strip=True)   # Extract text from the found div
    match = re.search(r'\bLeipzig\b', text)  # Search for specific place names
    if match:
        place_text = match.group(0)  # Extract the matched place text
else:
    place_text = ""  # Set place_text to empty string if the div is not found

# Extract the organization name from bounding box  "bbox 111 1497 978 1538"
div_org_name_tag = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 111 1497 978 1538" in value)

# Extract text from the third div with the specified format for orgName
if div_org_name_tag:
    org_name_text = ""
    text = div_org_name_tag.get_text(strip=True)   # Extract text from the found div
    match = re.search(r"B.G. Teubner", text)  # Search for a specific organization name format
    if match:
        org_name_text = match.group(0)  # Extract the matched organization name
else:
    org_name_text = ""  # Set org_name_text to empty string if the div is not found

# Extract the date text from bounding box "bbox 489 1387 737 1420"
div_date_tag = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 489 1387 737 1420" in value)

# Extract text from the second div with the specified format (\d+,\s\w+\s\d+)
if div_date_tag:
    date_text = ""
    text = div_date_tag.get_text(strip=True)   # Extract text from the found div
    match = re.search(r"\b\d{1,2},\s\w+\s\d{4}\b", text)  # Search for a specific date format
    if match:
        date_text = match.group(0)  # Extract the matched date text
else:
    date_text = ""  # Set date_text to empty string if the div is not found

# Find the div with class "ocr_carea" and p1 attribute containing "bbox 127 260 994 325"
div_p1 = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 127 260 994 325" in value)

# Extract text from the first div
if div_p1:
    p1 = div_p1.get_text(strip=True)   # Extract text from the found div
else:
    p1 = ""  # Set p1 to empty string if the div is not found

# Find the div with class "ocr_carea" and head1 attribute containing "bbox 126 378 674 425"
div_h1 = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 126 378 674 425" in value)

# Extract text from the first div
if div_h1:
    h1 = div_h1.get_text(strip=True)   # Extract text from the found div
else:
    h1 = ""  # Set h1 to empty string if the div is not found

# Define a list of target IDs
target_ids = ["bbox 127 413 992 465", "bbox 124 502 877 668"]

# Initialize a variable to store the extracted text
p2= ""

# Loop over each target ID
for target_id in target_ids:
    # Find the div element with the specified class and title
    div = soup.find("div", class_="ocr_carea", title=lambda value: value and target_id in value)

    # Extract text from the div if found
    if div:
        p2 += div.get_text(strip=True) + "\n"

# Find the div with class "ocr_carea" and head2 attribute containing "bbox 123 709 987 808"
div_h2 = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 123 709 987 808" in value)

# Extract text from the first div
if div_h2:
    h2 = div_h2.get_text(strip=True)   # Extract text from the found div
else:
    h2 = ""  # Set h2 to empty string if the div is not found

# Define a list of target IDs
target_ids = ["bbox 118 791 987 853", "bbox 122 826 986 879","bbox 123 859 512 894"]

# Initialize a variable to store the extracted text
p3= ""

# Loop over each target ID
for target_id in target_ids:
    # Find the div element with the specified class and title
    div = soup.find("div", class_="ocr_carea", title=lambda value: value and target_id in value)

    # Extract text from the div if found
    if div:
        p3 += div.get_text(strip=True) + "\n"

# Find the div with class "ocr_carea" and head3 attribute containing "bbox 120 911 277 949"
div_h3 = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 120 911 277 949" in value)

# Extract text from the first div
if div_h3:
    h3 = div_h3.get_text(strip=True)   # Extract text from the found div
else:
    h3 = ""  # Set h3 to empty string if the div is not found

# Define a list of target IDs
target_ids = ["bbox 120 960 985 1023", "bbox 115 1031 983 1088","bbox 111 1099 981 1273","bbox 109 1279 980 1448"]

# Initialize a variable to store the extracted text
p4= ""

# Loop over each target ID
for target_id in target_ids:
    # Find the div element with the specified class and title
    div = soup.find("div", class_="ocr_carea", title=lambda value: value and target_id in value)

    # Extract text from the div if found
    if div:
        p4 += div.get_text(strip=True) + "\n"

# Find the div with class "ocr_carea" and p5 attribute containing "bbox 168 110 759 323"
div_p5 = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 168 110 759 323" in value)

# Extract text from the first div
if div_p5:
    p5 = div_p5.get_text(strip=True)   # Extract text from the found div
else:
    p5 = ""  # Set p5 to empty string if the div is not found

# Find the div with class "ocr_carea" and head4 attribute containing "bbox 170 355 332 388"
div_h4 = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 170 355 332 388" in value)

# Extract text from the first div
if div_h4:
    h4 = div_h4.get_text(strip=True)   # Extract text from the found div
else:
    h4 = ""  # Set h4 to empty string if the div is not found

# Find the div with class "ocr_carea" and p6 attribute containing "bbox 172 405 474 439"
div_p6 = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 172 405 474 439" in value)

# Extract text from the first div
if div_p6:
    p6 = div_p6.get_text(strip=True)   # Extract text from the found div
else:
    p6 = ""  # Set p6 to empty string if the div is not found


# Construct the TEI XML document using the extracted information

# Define the TEI XML structure with extracted data
tei_xml = f"""
<TEI>
  <teiHeader>
    <fileDesc>
    <titleStmt>
        <title>{title}</title>
        </titleStmt>
        <orgName>{org_name_text}</orgName>
      <publicationStmt>
        <publPlace>{place_text}</publPlace>
        <date>{date_text}</date>
      </publicationStmt>
    </fileDesc>
  </teiHeader>
  <text>
    <body>
      <div>
        <p>{p1}</p>
         <head>{h1}</head>
        <p>{p2}</p>
         <head>{h2}</head>
         <p>{p3}</p>
         <head>{h3}</head>
          <p>{p4}</p>
      </div>
      <div>
        <p>{p5}</p>
        <head>{h4}</head>
        <p>{p6}</p>
      </div>
    </body>
  </text>
</TEI>
"""

# Write the TEI XML document to a file
with open("output_141813.tei.xml", "w", encoding="utf-8") as file:
    file.write(tei_xml)  # Write the TEI XML document content to the file
```
